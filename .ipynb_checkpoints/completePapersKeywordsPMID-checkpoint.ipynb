```python
import networkx as nx
import itertools
import matplotlib.pyplot as plt
import pandas as pd
import pickle
from pathlib import Path
```

```python
path_data = r'C:\Users\svalb\OneDrive\Documentos\GitHub\sci-abstract-graphs\completePapersKeywordsDataPMID.tsv'
path_graph = r'C:\Users\svalb\OneDrive\Documentos\GitHub\sci-abstract-graphs\graphCompletePapersKeywordsPMID.pickle'
```

```python
try:
    G = pickle.load(open(path_graph, 'rb'))
except FileNotFoundError:
    df=pd.read_csv(path_data, sep="\t")
    
    df.isna().value_counts()
    
    keywords_split = {}
    
    for el in range(0, len(df)):
        if(type(df.iloc[el]['MeshHeadings'])=='str'):
            keywords_split[df.iloc[el]['PMID']] = df.iloc[el]['MeshHeadings'].lower().split(',')
        elif(type(df.iloc[el]['MeshHeadings'])!='str'):
            keywords_split[df.iloc[el]['PMID']] = str(df.iloc[el]['MeshHeadings']).lower().split(',')
    
    for PMID in keywords_split.keys():
        for keyword in range(0, len(keywords_split[PMID])):
            keywords_split[PMID][keyword] = keywords_split[PMID][keyword].strip()
            
    G = nx.Graph()
    
    for PMID_value in keywords_split:
        for combination in itertools.combinations(keywords_split[PMID_value], 2):
            if combination[0] != combination[1]:
                if G.has_edge(combination[0], combination[1]) or G.has_edge(combination[1], combination[0]):                
                    G[combination[0]][combination[1]]['weight'] += 1
                    G[combination[0]][combination[1]]['PMID'].append(PMID_value)
                else:
                    G.add_edge(combination[0], combination[1], weight = 1, PMID = [PMID_value])
                    
    pickle.dump(G, open(path_graph, 'wb'))
                    
    threshold_weight = 150000

    nodes_to_keep = set()
    for u, v, data in G.edges(data=True):
        if data['weight'] > threshold_weight:
            nodes_to_keep.update([u, v])
```

```python
nodes_to_keep
```

```python
H = G.subgraph(nodes_to_keep)
```

```python
elarge_3 = [(u, v) for (u, v, d) in H.edges(data=True) if d["weight"] > 150000]
esmall_3 = [(u, v) for (u, v, d) in H.edges(data=True) if d["weight"] <= 150000 and d["weight"] >100000]
edge_labels = nx.get_edge_attributes(H, "weight")
edge_labels_2 = edge_labels.copy()
for key in edge_labels.keys():
    if edge_labels[key] <= 100000:
        edge_labels_2.pop(key)
```

```python
plt.figure(figsize=(16, 8))
pos = nx.spring_layout(H, seed=10)

nx.draw_networkx_nodes(H, pos)
nx.draw_networkx_edges(H, pos, edgelist=elarge_3, width=2)
nx.draw_networkx_edges(H, pos, edgelist=esmall_3, width=1, style="dashed")

nx.draw_networkx_labels(H, pos, font_size=16, font_color="r", font_family="sans-serif")
# edge weight labels

nx.draw_networkx_edge_labels(H, pos, edge_labels_2, font_size=12, font_color="r")

ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.tight_layout()
plt.show()
```
