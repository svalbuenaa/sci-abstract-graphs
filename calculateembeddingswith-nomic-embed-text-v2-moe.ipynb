```python
model = SentenceTransformer('nomic-ai/nomic-embed-text-v2-moe', trust_remote_code=True)

df = pd.read_csv("/kaggle/input/input-embeddings-21814843-to-22570866/DFsPreparedForEmbedding_21814843_to_22570866.csv")
```

```python
abstracts = []

for el in range(0, len(df.Abstract)):
    abstracts.append(df.Abstract[el])
```

```python
embeddings = []
for el in tqdm(range(len(abstracts))):
    embeddings.append(model.encode(abstracts[el], show_progress_bar = False))
```

```python
outputs = []
for el in embeddings:
    outputs.append(list(el))

df["AbstractEmbedding"] = outputs 

df.to_csv("/kaggle/working/DFsEmbedding_21814843_to_22570866_nomic-embed-text-v2-moe.csv")
```
